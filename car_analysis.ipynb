```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
car_df=pd.read_csv('car data.csv')
car_df.info()
```

```python
car_df.isnull().sum()
```

```python
numeric_columns = car_df.select_dtypes(include=['number'])
# Calculate mean, median, standard deviation, minimum, and maximum
summary_stats = numeric_columns.agg(['mean', 'median', 'std', 'min', 'max'])
print(summary_stats)
```

```python
car_df.duplicated().sum()
```

```python
car_df = car_df.drop_duplicates()
car_df.duplicated().sum()
```

```python
numeric_columns = car_df.select_dtypes(include=['number']).columns #exclude non-numeric dtypes
corr = car_df[numeric_columns].corr()
print(corr)
```

```python
plt.figure(figsize=(10,4))
sns.heatmap(corr,annot = True,fmt=".2f",cmap='hot')
plt.show()
```

```python
car_df.info()
```

```python
numerical_features = ['Year', 'Driven_kms', 'Selling_Price', 'Present_Price','Owner']
for feature in numerical_features:
    plt.figure(figsize=(4,4))
    sns.displot(data=car_df, x=feature, kde=True)
    plt.title(f'Distribution of {feature}')
    plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.scatterplot(x='Driven_kms', y='Selling_Price', data=car_df)
plt.title('Relationship between Driven_kms and Selling_Price')
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.scatterplot(x='Present_Price', y='Selling_Price', data=car_df)
plt.title('Relationship between Driven_kms and Selling_Price')
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.scatterplot(x='Driven_kms', y='Present_Price', data=car_df)
plt.title('Relationship between Driven_kms and Selling_Price')
plt.show()
```
